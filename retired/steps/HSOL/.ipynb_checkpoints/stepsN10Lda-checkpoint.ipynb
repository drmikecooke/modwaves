```python
N,L_d=10,16
import sys

# setting path
sys.path.append('/home/mike/Documents/github/modwaves')

import matplotlib.pyplot as plt
import numpy as np

from scipy.interpolate import make_interp_spline as mis

from waterwaves1.wave import wave1,wave_base,Hmx_d
from waterwaves1.display import eta

def check_base(SOL):
    success=[sol.success for sol in SOL]
    return np.all(success)

def extract(SOL,attr):
    return [getattr(sol,attr) for sol in SOL]
```

```python
def dspEk(N,H,L_d):
    SOL=wave_base(N,H,L_d)
    SOLx=np.array(extract(SOL,'x'))
    plt.plot(H,SOLx[:,:-2])

dspEk(10,np.linspace(0,1),L_d) # check breakdown
```

```python
H=np.linspace(0,0.6,4)
SOL=wave_base(N,H,L_d) # get preliminary base
print(check_base(SOL))
SOLx=np.array(extract(SOL,'x'))

fs=mis(H,SOLx,k=min(len(H)-1,3))
Hp=np.linspace(0,1)
plt.plot(Hp,fs(Hp)[:,:-2])
plt.show()
```

```python
H=np.linspace(0,0.9,8)
SOL=wave_base(N,H,L_d,fs=fs) # refine base
SOLx=np.array(extract(SOL,'x'))
print(check_base(SOL))
fs=mis(H,SOLx,k=min(len(H)-1,3))
plt.plot(Hp,fs(Hp)[:,:-2])
plt.show()
```

```python
plt.plot(Hp,fs(Hp)[:,-2:],label=['Q','R'])
plt.legend()
plt.show()
```

```python
X=np.linspace(0,np.pi,200)

for h in np.linspace(0.05,0.95,4):
    d=2*np.pi/L_d
    solx=fs(h)
    Np=40
    A0=np.insert(solx,-2,np.zeros(Np))
    solNp=wave1(N+Np,h*Hmx_d(L_d)*d,d=d,A0=A0)
    if not solNp.success:
        print(solNp)
    eta(X,d,solNp,f'h={h:4.2f}')
plt.legend()
plt.show()
```
