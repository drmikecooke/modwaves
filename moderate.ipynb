```python
import matplotlib.pyplot as plt
import numpy as np

from waterwaves import unpack,IEj,wave
```

```python
# odd n
X=np.linspace(0,np.pi,100)
N,d=20,2*np.pi
for n in range(1,14,2): # step two at time so graph not too busy
    H=0.02*n*np.pi
    sol=wave(N,H)
    if not sol.success: # weed out fails and print analysis
        print(sol)
        continue
    Ej,Bj,Ub,Q,R=unpack(sol.x)
    plt.plot(X,IEj([d,*Ej],X),label=f"Num: H/d={H/2/np.pi:6.4f}")
plt.legend()
plt.title(f'Test odd d={d:6.4f}')
plt.show()
```

```python
# even n
for n in range(2,14,2): # step two at time so graph not too busy
    H=0.02*n*np.pi
    sol=wave(N,H)
    if not sol.success: # weed out fails and print analysis
        print(sol,H)
        continue
    Ej,Bj,Ub,Q,R=unpack(sol.x)
    plt.plot(X,IEj([d,*Ej],X),label=f"Num: H/d={H/2/np.pi:6.4f}")
plt.legend()
plt.title(f'Test odd d={d:6.4f}')
plt.show()
```
