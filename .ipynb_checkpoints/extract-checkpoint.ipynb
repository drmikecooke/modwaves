```python
import matplotlib.pyplot as plt
import numpy as np

from waterwaves1.wave import wave1
from waterwaves1.packing import unpack1
from waterwaves1.fourier import IEj
```

```python
L_d=12
Hmx_d=(0.141063*L_d+0.0095721*L_d**2+0.0077829*L_d**3)/(1+0.0788340*L_d+0.0317567*L_d**2+0.0093407*L_d**3)
d=2*np.pi/L_d
N,h,s=40,0.8,d*(1+0.0*Hmx_d)
```

```python
sol1=wave1(N,h*Hmx_d*d,d=d,scaler=s)
Ej1,Q1,R1=unpack1(sol1.x)
sol1.success
```

```python
X=np.linspace(0,np.pi)
plt.plot(X,IEj([d,*Ej1],X),label=f"Ej1")
plt.legend()
```

```python
sol2=wave1(N,0.96*Hmx_d*d,d=d,scaler=s,A0=sol1.x)
Ej2,Q2,R2=unpack1(sol2.x)
plt.plot(X,IEj([d,*Ej2],X),label=f"Ej2")
plt.legend(),sol2.success
```

```python
sol3=wave1(N,0.97*Hmx_d*d,d=d,scaler=s,A0=sol2.x)
Ej3,Q3,R3=unpack1(sol3.x)
plt.plot(X,IEj([d,*Ej3],X),label=f"Ej3")
plt.legend(),sol3.success
```

```python
sol4=wave1(N,0.98*Hmx_d*d,d=d,scaler=s,A0=sol1.x)
Ej4,Q4,R4=unpack1(sol4.x)
plt.plot(X,IEj([d,*Ej4],X),label=f"Ej4")
plt.legend(),sol4.success
```

```python
sol5=wave1(N,0.987*Hmx_d*d,d=d,scaler=s,A0=sol4.x)
Ej5,Q5,R5=unpack1(sol5.x)
plt.plot(X,IEj([d,*Ej5],X),label=f"Ej5")
plt.legend(),sol5.success
```

```python
from waterwaves1.grids import grids
```

```python
ss=np.linspace(-0.15,0.5)
SS=d*(1+ss*Hmx_d)
Cnd=[]
for s in SS:
    g=grids(N,s)
    
    N=len(Ej1)
    Etam=IEj([d,*Ej1],g.Xm) # imposes depth constraint
    Etamj=Etam.reshape((N+1,1))@g.j.reshape((1,N))
    C,S=(np.exp,np.exp) if d==0 else (np.cosh,np.sinh) # hyperbolics for Y dependence, for deep water use exp
    CEmj,SEmj=C(Etamj)/g.scl,S(Etamj)/g.scl
    a=np.hstack([-Etam.reshape(N+1,1),(SEmj*g.CXmj)])
        
    Cnd+=[np.linalg.cond(a)]

plt.plot(ss,Cnd)
```
