# QR trends

```python
import matplotlib.pyplot as plt
import numpy as np

from waterwaves1.wave import wave1
from waterwaves1.packing import unpack1,pack1
from waterwaves1.fourier import IEj
```

```python
L_d=8
Hmx_d=(0.141063*L_d+0.0095721*L_d**2+0.0077829*L_d**3)/(1+0.0788340*L_d+0.0317567*L_d**2+0.0093407*L_d**3)
d=2*np.pi/L_d
N,s=40,d*(1+0.0*Hmx_d)
X=np.linspace(0,np.pi)
```

```python
def Qplot(H):
    L=[]
    for h in H:
        sol=wave1(N,h*Hmx_d*d,d=d,scaler=s)
        Ej,Q,R=unpack1(sol.x)
        if not sol.success:
            print(h,'fail')
        L+=[Q]
    plt.plot(H,L)

Qplot(np.linspace(0,0.6))
```

```python
def Rplot(H):
    L=[]
    for h in H:
        sol=wave1(N,h*Hmx_d*d,d=d,scaler=s)
        Ej,Q,R=unpack1(sol.x)
        if not sol.success:
            print(h,'fail')
        L+=[R]
    plt.plot(H,L)
Rplot(np.linspace(0,0.6))
```
